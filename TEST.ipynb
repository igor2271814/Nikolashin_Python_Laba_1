Гистограмма бросков двух кубиков d20

```python
import pandas as pd
import numpy as np


data = pd.read_csv('Airpollution.csv')# Чтение данных

data_number = data.select_dtypes(include=[np.number]) # Выбор только числовых данных
data_log = pd.DataFrame() # Создание нового датафрейма для хранения логарифмированных данных
# Цикл по столбцам числовых данных
for column in data_number.columns:
    data_log[column] = np.log(data_number[column] + 1)  # Логарифмирование после прибавления 1
print(data_log)
```
